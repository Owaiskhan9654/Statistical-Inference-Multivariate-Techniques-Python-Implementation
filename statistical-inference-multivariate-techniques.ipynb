# **Z Score Outlier Detection using Z Score**

### Outlier is a data point in a data set that is too far from all other observation
Those data point which falls ouside 3 standard deviations can be classified as outliers

```python
import math
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from scipy.special import ndtr as ndtr
```

#### Let us consider our dataset of Normal Distribution having Mean around 50 and standard deviation of 10 . we will take 100 such points randomly

```python
mu,sigma=50,10
input=np.random.normal(mu,sigma,100)
```

Let us take some Outliers in our dataset which we will be finding again using function defined

```python
input[99]=350
input[3]=-200
input[45]=5000
```


**Formula for Zscore**

z=(observed value - mean) / Standard Deviation

z= ( X - μ ) / σ

```python
outlier_list=[]

z_score_list=[]
def outliers(input):
    threshold=3 #Selecting threshold as 3 stadard deviation
    mean=np.mean(input)
    standard_deviaton=np.std(input)

    for i in input:
        z_score= (i-mean)/standard_deviaton
        z_score_list.append(z_score)
        if np.abs(z_score)>threshold:
            outlier_list.append(i)
    return outlier_list       
```

```python
p=outliers(input)
```

```python
print('outliers present are ',p)
```

```python
print(z_score_list)
```

Now we will find outaliers using Alpha Value . For that we have find standardize variable

```python
def outliers_alpha(alpha):
    p_values_list=(ndtr(np.abs(z_score_list)))
    outlier_list1=(p_values_list>alpha).astype(int)
    print('Those values which are outlier are marked as 1')
    for i in zip(input,outlier_list1):
        print(i)
```

```python
outliers_alpha(0.95)
```

```python
  def t_dist(input1=input1,input_sample_size=80000):
     
     input_sample=np.random.choice(input1,input_sample_size)
     mean_population=np.mean(input1)
     mean_sample=np.mean(input_sample)
     s=np.std(input_sample)
     dof=input_sample_size-1

     t_score= (mean_sample-mean_population)/(s/np.sqrt(input_sample_size-1))
     return t_score,dof
```

```python
tscore,df=t_dist(input1)
print(tscore)
```

# **Significance of F Score**

```python
mu1,sigma1=20,5
mu2,sigma2=10,2
input1=np.random.normal(mu1,sigma1,20)
input2=np.random.normal(mu2,sigma2,10)
```

```python
var1=np.var(input1)
var2=np.var(input2)
```

```python
fstatistics=var1/var2
```

```python
w=fstatistics
dof1=len(input1)-1
dof2=len(input2)-1
```

```python
def fstas(w=w,dof1=dof1,dof2=dof2):
    f=(math.gamma(dof1+dof2)/2)*((dof1/dof2)**(dof1/2))*(w**((dof1/2)-1))/((math.gamma(dof1/2)*math.gamma(dof2/2))*(((dof1/dof2)*(w+1))**(dof1+dof2)/2))
    return f
```

# **Significance of Chi Square Test**
Now doing same procedure as we did in Zscore calculation

```python
mu,sigma=50,10
input1=np.random.normal(mu,sigma,100)
```

```python
def z_square(input):
    z_score_square_sum=0
    mean=np.mean(input)
    standard_deviaton=np.std(input)

    for i in input:
        z_score= (i-mean)/standard_deviaton
        z_score_square=z_score**2
        z_score_square_sum=z_score_square_sum + z_score_square
    
    return z_score_square_sum
```
